```python
import scipy.stats as stats
import numpy as np

# Observed frequencies (example data)
observed_frequencies = np.array([22, 18, 19, 8, 11, 22])  # blue, orange, green, yellow, red, brown

# Expected frequencies based on the given proportions
total_m_and_ms = observed_frequencies.sum()
expected_proportions = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])
expected_frequencies = expected_proportions * total_m_and_ms

# Perform the chi-square goodness of fit test
chi_square_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

chi_square_statistic, p_value
```

```python
# Given values
sample_size = 500
number_of_smokers = 60

# Calculate the sample proportion
sample_proportion = number_of_smokers / sample_size

# Calculate the standard error for the proportion
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Find the z-value for the 95% confidence level
z_value = stats.norm.ppf(0.975)  # Two-tailed, so 0.975 for 95% confidence

# Calculate the margin of error
margin_of_error = z_value * standard_error

# Calculate the confidence interval
lower_limit = sample_proportion - margin_of_error
upper_limit = sample_proportion + margin_of_error

lower_limit, upper_limit, margin_of_error
```

```python
import scipy.stats as stats

# Given values
sample_mean = 75
sample_std = 12
sample_size = 100  # Assuming a sample size of 100 for this example
confidence_level = 0.90

# Calculate the standard error
standard_error = sample_std / (sample_size ** 0.5)

# Find the z-value for the 90% confidence level
z_value = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_value * standard_error

# Calculate the confidence interval
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

lower_limit, upper_limit, margin_of_error
```

```python
import numpy as np
from scipy.stats import norm

# Sample size
n = 1000

# Sample proportion preferring Coke
p_hat = 520 / n

# Confidence level (1 - alpha) for 99% CI
alpha = 1 - 0.99

# Critical value from standard normal distribution (z-score)
z_crit = norm.ppf(1 - alpha / 2)  # Using ppf for two-tailed interval

# Standard error of the proportion
se = np.sqrt(p_hat * (1 - p_hat) / n)

# Confidence interval lower and upper bounds
lower_bound = p_hat - z_crit * se
upper_bound = p_hat + z_crit * se

print("99% Confidence Interval for Proportion Preferring Coke:")
print(f"Lower Bound: {lower_bound:.4f}")
print(f"Upper Bound: {upper_bound:.4f}")
```

```python
# 1. Define Hypotheses:

# Null Hypothesis (H0): The coin is fair. In other words, the probability of heads (p_heads) and tails (p_tails) are equal (p_heads = p_tails = 0.5).
# Alternative Hypothesis (H1): The coin is biased towards tails. This means p_tails > p_heads.
# 2. Expected Frequencies:

# Since we assume a fair coin (H0), the expected frequency for both heads and tails outcomes in 100 flips is:

# Expected Heads (E_heads) = 0.5 * 100 = 50
# Expected Tails (E_tails) = 0.5 * 100 = 50
# 3. Observed Frequencies:

# From the experiment, we have:

# Observed Tails = 45
# Observed Heads = 100 - Observed Tails = 100 - 45 = 55 (calculated)
# 4. Chi-Square Calculation:

# We use the chi-square formula for goodness of fit:

#  =  ( (Observed - Expected) / Expected )
# where:

#  (chi-square) is the test statistic
#  (summation) is taken over all categories (heads and tails in this case)
# Observed is the observed frequency for each category
# Expected is the expected frequency for each category
# Calculate chi-square:

#  = ( (55 - 50) / 50 ) + ( (45 - 50) / 50 )
#  = ( 25 / 50 ) + ( 25 / 50 )
#  = 0.5 + 0.5
#  = 1.0
# 5. Degrees of Freedom:

# The degrees of freedom (df) for a chi-square goodness of fit test with k categories is:

# df = k - 1
# Here, k = 2 (heads and tails). Therefore, df = 2 - 1 = 1

# 6. Decision Rule:

# We will use a significance level () of 0.05. Look up the chi-square critical value with df = 1 and  = 0.05 in a chi-square distribution table. You can find this online or in statistics textbooks.

# Critical value (_critical)  3.84

# 7. Conclusion:

# If the calculated chi-square () is greater than the critical value (_critical), we reject the null hypothesis (H0).
# If the calculated chi-square () is less than or equal to the critical value (_critical), we fail to reject the null hypothesis (H0).
# In this case,  (1.0) is less than _critical (3.84). Therefore, based on this test at a significance level of 0.05, we fail to reject the null hypothesis. We cannot conclude with sufficient evidence that the coin is biased towards tails. There is a chance this observed difference arose by random chance for a fair coin.
```
